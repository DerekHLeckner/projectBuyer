```python
def screenCaptureSetup():
    sleep(1)
    try:
        driver.switch_to.window(driver.current_url)
    except:
        pass
    sleep(1)
    pyautogui.moveTo(10, 250, 2)   
    pyautogui.click(10, 250)
    sleep(1)
    keyboard.press_and_release('alt+shift+p')
    sleep(5)
    while len(driver.window_handles) != 2:
        print(len(driver.window_handles))
        sleep(2)
    driver.switch_to.window(driver.window_handles[1])
    try:
        driver.find_element_by_xpath("//a[@class='close close-topright']").click()
        sleep(1)
    except:
        pass
    pyautogui.click(10, 250, button ='right')
    sleep(1)
    driver.find_element_by_xpath("//span[@class='lhs']").click()
    sleep(2)
    keyboard.press_and_release("tab")
    sleep(2)
    keyboard.press_and_release("enter")
    sleep(3)
    keyboard.press_and_release("esc")
    sleep(1)
    keyboard.press_and_release('ctrl+w')
    sleep(1)
    driver.switch_to.window(driver.window_handles[0])
```

```python
def screenshot():
    try:
        sleep(1)
#         pyautogui.click(13, 250)
        keyboard.press_and_release('alt+shift+p')
        driver.switch_to.window(driver.window_handles[1])
    except:
        gotdata = False
        while gotdata == False:
            try:
                driver.switch_to.window(driver.window_handles[1])
                gotdata = True
            except:
                sleep(1)
            try:
                driver.find_element_by_xpath("//a[@class='close close-topright']").click()
            except:
                empty_variable = " "
                
                sleep(1)
```

```python
def desktop_scroll():
    while driver.find_element_by_xpath("//button[@class='load-more button row']").text == 'LOAD MORE':
        sleep(3)
        pyautogui.moveTo(500, 800)
        sleep(1)
        pyautogui.scroll(-600)
        sleep(1)
        driver.find_element_by_xpath("//button[@class='load-more button row']").click()
    
    for i in range(3):
        pyautogui.scroll(+300)
```

```python
#Load Data
agents_df = pd.read_csv('Resources/agents.csv', index_col=0)
query_df = pd.read_csv('Resources/queryList.csv', index_col=0)
unitsCount = pd.read_csv('Resources/unitsCount.csv', index_col=0)
available_units = pd.read_csv("Resources/availableUnits.csv", index_col=0)
todaysDates = datetime.now().strftime("%m-%d-%y")

# Do not run

available_dic = {}
attributes = ['company','building', 'date', '0', '1', '2', '3', '4', '5']
for attr in attributes:
    available_dic[attr] = []
# Do not run

units_dic = {}
attributes = ['company','building','date', 'unitName', 'sqrFt','price','availableDate','bdr','bthr','flr','features','apt_views','hasSpecial']#,'specialDetails'
for attr in attributes:
    units_dic[attr] = []
########################################################################

# Checks to see if there are still buildings to query
building_list = query_df['buildingUrl'].tolist()
number_buildings = len(building_list)
while len(building_list) > 0: 
    print(f"There are {len(building_list)} buildings left to query")
    print(' ')
    
    # Load agents
    agents_list = agents_df['agents'].tolist()
    number_agents = len(agents_list)
    print(f"Loading {number_agents} agents")
    print(' ')
    #Select agent
    for a in range(number_agents):
        if a == 0: # Default at beginning of step, an agent at random will be selected and given the ability to pass to the mission
            agentContinues = 1
            selectedAgent = random.choice(agents_list)
            agents_list.remove(selectedAgent)
            print(f"Agent {selectedAgent} was selected")
            print(" ")  
            # Choose cookie directory

            directory = agents_df[agents_df['agents']==selectedAgent]['cookieDirectory'].tolist()[0]

            # Load cookies
            cookieFile = open(directory,"rb")
            savedCookies = pickle.load(cookieFile)
            cookieBig = savedCookies
            # Execute driver
            options = Options()
            #                 options.add_argument(r"--user-data-dir=C:\path\to\chrome\user\data") #e.g. C:\Users\You\AppData\Local\Google\Chrome\User Data
            #                 options.add_argument(r'--profile-directory=YourProfileDir') #e.g. Profile 3
            options.add_argument(f"user-agent={selectedAgent}")
            options.add_extension('Resources/screenCapture.crx')
            driver = webdriver.Chrome(options=options)
            sleep(1)
            driver.switch_to.window(driver.window_handles[1])
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            sleep(1)
            driver.maximize_window()
            sleep(2)
            driver.get('https://www.hellofresh.com/')
            sleep(3)
#             screenCaptureSetup()
            for cookie in savedCookies:
                try:
                    driver.add_cookie(cookie)
                except:
                    pass
      
        else:
            agentContinues = random.randrange(0,2)
            print(f"Agent has gotten a continue order of {agentContinues}")
            if agentContinues == 0:

                # If agent does not continue, session closes, a new agent is selected
                try:
                    pickle.dump(cookieBig, open(directory,"wb"))
                    cookieFile.close()   
                    driver.close()
                except:
                    pass
                
                print(f"This agent was set aside with an order of {agentContinues}")
                selectedAgent = random.choice(agents_list)
                agents_list.remove(selectedAgent)
                print(f"{selectedAgent} is the new agent")
                # Once a new agent is selected, they are put in place and it happens all over again
                # Choose cookie directory
                directory = agents_df[agents_df['agents']==selectedAgent]['cookieDirectory'].tolist()[0]
                # Load cookies
                cookieFile = open(directory,"rb")
                savedCookies = pickle.load(cookieFile)
                cookieBig = savedCookies
                # Execute driver
                options = Options()
                #                 options.add_argument(r"--user-data-dir=C:\path\to\chrome\user\data") #e.g. C:\Users\You\AppData\Local\Google\Chrome\User Data
                #                 options.add_argument(r'--profile-directory=YourProfileDir') #e.g. Profile 3
                options.add_argument(f"user-agent={selectedAgent}")
                options.add_extension('Resources/screenCapture.crx')
                driver = webdriver.Chrome(options=options)
                sleep(1)
                driver.switch_to.window(driver.window_handles[1])
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
                sleep(1)
                driver.get('https://www.hellofresh.com/')
                sleep(5)
                driver.maximize_window()
#                 screenCaptureSetup()
                for cookie in savedCookies:
                    try:
                        driver.add_cookie(cookie)
                    except:
                        pass
    
        while (agentContinues == 1) & (len(building_list) > 0):
            selectedBuilding= random.choice(building_list)
            print(f"Querying building {selectedBuilding}")
# Do not run


# This rule is attempted to be uniform, however chelsea is weird so an exception must be made
            if 'https://www.udr.com/new-york-city-apartments/chelsea/21-chelsea/apartments-pricing' not in selectedBuilding:
                maxBedroom = query_df[query_df['buildingUrl'] == selectedBuilding]['maxBdr']
                bedroomQueries = [i+1 for i in range(int(maxBedroom))]
                bedroomQueries.append(0)
                number_bdr_queries = len(bedroomQueries)
            else:
                bedroomQueries = ["0",'1','2','5']
                number_bdr_queries = len(bedroomQueries)
              # Do not run
# Now to cover all bedrooms in one building
            for bdr in range(number_bdr_queries):
                selectedBdr = random.choice(bedroomQueries)
                bedroomQueries.remove(selectedBdr)
                queryUrl = f"{selectedBuilding}?beds={selectedBdr}"
                driver.get(queryUrl) # We get the url to query
                newCookies = driver.get_cookies()
                cookieBig = cookieBig+newCookies
                sleep(13)
                
                # Get agent index
                agent_index = agents_df[agents_df['agents'] == selectedAgent].index[0]
                
                # Get building index
                query_index = query_df[query_df['buildingUrl'] == selectedBuilding].index[0]                
                
                # UPDATES
                buildingName = query_df.at[query_index, 'building'].replace(" ","")
                queryingCompany = query_df.at[query_index, 'company']
                agents_df.at[agent_index, "missionCount"]+=1 # Update Mission
                agents_df['lastMissionDate'].iloc[agent_index] = datetime.now().strftime("%m-%d-%y")                  
                agents_df['lastMissionBuilding'].iloc[agent_index] = buildingName # update building
                agents_df['lastMissionCompany'].iloc[agent_index] = queryingCompany # update company
                
                # Load Full List
                desktop_scroll()
                sleep(1)
                
                # Grab available beds
                amount_bdrs = int(driver.find_element_by_xpath("//span[@class='amount']").text)
                available_dic[str(selectedBdr)].append(amount_bdrs)
                
                # Grab unit information
                if amount_bdrs > 0:
#                     screenshot()
                    sleep(1)
                    try:
                        driver.find_element_by_xpath("//a[@class='close close-topright']").click()
                    except:
                        pass
#                     save_desktop_screenshot()
                    parent = driver.find_element_by_xpath('//ul[@id="unitResults"]')
                    
                    holder = [units_dic['sqrFt'].append(re.findall(r'\d+', sqrFt.text)[0]) for sqrFt in driver.find_elements_by_xpath('//li[@class ="hide-for-small sqft"]')if len(sqrFt.text) > 0]
                    holder = [units_dic['price'].append(re.findall(r'\d+', price.text.replace(',', ""))[0]) for price in driver.find_elements_by_xpath('//li[@class ="price"]') if (len(price.text) > 0) and (":" in price.text)]
                    holder = [units_dic['availableDate'].append(available_date.text.split('\n')[1]) for available_date in driver.find_elements_by_xpath('//li[@class ="available"]')if len(available_date.text) > 0]
                    holder = [units_dic['apt_views'].append(view.get_attribute("data-views")) for view in driver.find_elements_by_xpath('//li[@data-views]')]
                    holder = [units_dic['features'].append(additional_features.text.replace('Apartment Features\n',"")) for additional_features in driver.find_elements_by_xpath('//li[@class ="features restricted"]')if len(additional_features.text) > 0]
                    holder = [units_dic['flr'].append(floor.text) for floor in driver.find_elements_by_xpath('//span[@class="floor-number"]') if len(floor.text)>0]
                    holder = [units_dic['bdr'].append(selectedBdr) for bed in driver.find_elements_by_xpath('//a[@class ="unit-name-expand hide-for-small"]')]
                    holder = [units_dic['bthr'].append(bath.split(' ')[0]) for bath in re.findall(r'\d?\.?\d Bathroom', parent.text)]
    
                    for listing in driver.find_elements_by_xpath('//a[@class ="unit-name-expand hide-for-small"]'):
                        if len(listing.text) > 0:
                            units_dic['unitName'].append(listing.text) 
                            units_dic['date'].append(datetime.now().strftime("%m-%d-%y"))
                            units_dic['building'].append(query_df.at[query_index, 'building'])
                            units_dic['company'].append("UDR")


                    
                    #Special Offer code
                    actions = driver.find_elements_by_xpath('//li[@class="actions desktop"]')
                    expand = driver.find_elements_by_xpath('//button[@aria-label="Apartment Details"]')

                    for z in range(len(actions)):
                        if 'Special Offer' in actions[z].text:
                            units_dic['hasSpecial'].append(1)
#                             expand[z].click()
#                             sleep(3)
#                             units_dic['specialDetails'].append(driver.find_element_by_xpath('//li[@class="special-offers hide-for-small"]').text.split('\n')[1])
                            
#                             expand[z].click()
#                             sleep(3)
                        else:
                            units_dic['hasSpecial'].append(0)
#                             units_dic['specialDetails'].append('No Offers')
                else:
                    pass
                # Bedrooms finished
            print("Bedrooms finished, building done")
# Do not run
            building_list.remove(selectedBuilding)
            available_dic['date'].append(datetime.now().strftime("%m-%d-%y"))
            available_dic['building'].append(query_df[query_df['buildingUrl'] == selectedBuilding]['building'].tolist()[0])
            available_dic['company'].append('UDR')
            for available_rooms in available_dic.keys():
                if len(available_dic[available_rooms]) != len(available_dic['date']):
                    available_dic[available_rooms].append(np.nan)
            agentContinues = random.randrange(0,2)
        if len(building_list) == 0:
            break
# Do not run
# Merge new data with old
crawledUnits_df = pd.DataFrame.from_dict(units_dic)
crawledCounts_df = pd.DataFrame.from_dict(available_dic)

updated_available_units = pd.concat([available_units, crawledUnits_df])
updated_count_units = pd.concat([unitsCount, crawledCounts_df])
print("Saving updated files starting")
updated_count_units.to_csv('Resources/unitsCount.csv')
updated_available_units.to_csv("Resources/availableUnits.csv")
agents_df.to_csv("Resources/agents.csv")
print("Saving updated files finished")
driver.close()
```
